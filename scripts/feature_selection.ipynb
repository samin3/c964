```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# sample correlation of features in a random stock
df = pd.read_csv('../raw/AMZN_2006-01-01_to_2018-01-01.csv')
df = df[['Open','High','Low','Close','Volume']]
plt.subplots()
plt.pcolor(df.corr())
plt.suptitle('Feature Correlation')
plt.yticks(np.arange(0.5,len(df.columns),1), df.columns)
plt.xticks(np.arange(0.5,len(df.columns),1), df.columns, rotation=90)
plt.colorbar()
plt.show()
```
